```python
train_df= pd.read_csv('/content/drive/MyDrive/new ML-DL/titanic/train.csv')
train_df.head()
```

```python
df= pd.concat([test_df,train_df],axis=0).reset_index(drop=True)
df
```

```python
df=df.drop(['PassengerId','Name','Cabin','Ticket'],axis=1)
```

```python
df['Sex']=(df['Sex']== 'male').astype(int)
```

```python
le=LabelEncoder()
df['Embarked']=le.fit_transform(df['Embarked'])
df
```

```python
def standered_scaler(dataFrame,oversample=False):
  X= dataFrame[dataFrame.columns[:-1]].values
  y= dataFrame[dataFrame.columns[-1]].values

  scaler= StandardScaler()
  X=scaler.fit_transform(X)
  if oversample:
    ros=RandomOverSampler()
    X,y=ros.fit_resample(X,y)

  data= np.hstack((X,np.reshape(y,(-1,1))))

  return data,X,y
```

```python
print(len(df))
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
```

```python
df.isnull().sum()
```

```python
train, valid,test = np.split(df.sample(frac=1),[int(0.6*len(df)), int(0.8*len(df))])

valid
```

```python
print(len(train[train['Survived']==1])) # gamma
print(len(train[train['Survived']==0]))
```

```python
data,x_train,y_train= standered_scaler(dataFrame=train,oversample=True)
data,x_valid,y_valid= standered_scaler(dataFrame=valid,oversample=False)
data,x_test,y_test= standered_scaler(dataFrame=test,oversample=False)
```

```python
model_lr= LogisticRegression()
history=model_lr.fit(x_train,y_train)
```

```python
y_predict=model_lr.predict(x_test)

# y_predict
```

- overfitting checking

```python
train_predict= model_lr.predict(x_train)
train_accuracy= accuracy_score(y_train,train_predict)

test_preds = model_lr.predict(x_test)
test_acc = accuracy_score(y_test, test_preds)


print('Train Accuracy : \n', train_accuracy)
print('Test Accuracy : \n', test_acc)
```

```python
print('Accuracy Report \n',accuracy_score(y_test,y_predict))
print('Classification Report: \n',classification_report(y_test,y_predict))
```
